## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10));

# 3. Plot data
ax.plot(x, y);

# 4. Customize plot
ax.set(title="My first plot", xlabel="x-axis", ylabel="y-axis");

# 5. Save and show
fig.savefig("my_first_plot.png");
plt.show()
```

## Making figures with Numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
#Create some data
x = np.linspace(0, 10, 100)
```

```python
# Plot the data
fig, ax = plt.subplots();
ax.plot(x, x**2);
plt.show()
```

```python
# Use same data to make scatter
fig, ax = plt.subplots();
ax.scatter(x,np.exp(x));
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots();
ax.scatter(x, np.sin(x));
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, "Peanut butter": 8, "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut butter prices", ylabel="Price ($)", xlabel="Nut butter type")
plt.show()
```
